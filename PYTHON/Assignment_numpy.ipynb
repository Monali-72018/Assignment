```python
# 19. Question: Create a random 4x4 matrix and use NumPy to compute its eigenvalues and eigenvectors. Verify the eigenvalues by reconstructing the matrix.

array = np.random.randint(0,50, size = (4,4))
print("Original Array:\n", array)

eigenvalues, eigenvectors = np.linalg.eig(array)

print("\nEigenvalues:\n", eigenvalues)
print("\nEigenvectors:\n", eigenvectors)

D = np.diag(eigenvalues)

array_reconstructed = eigenvectors @ D @ np.linalg.inv(eigenvectors)

print("\nReconstructed Matrix:\n", array_reconstructed)
print("\nVerification Result:", np.allclose(array, array_reconstructed))
```

```python
# 21. Question: Perform matrix multiplication of two 2D arrays using np.dot () and @. Compare the results and performance of both methods using a large dataset.
import time

A = np.random.rand(1000, 1000)
B = np.random.rand(1000, 1000)
print("A: ",A)
print("B: ",B)


start = time.time()
matrix_multiplication = np.dot(A,B)
end = time.time()
print("Time taken using np.dot():", end - start)

start = time.time()
result_at = A @ B
end = time.time()
print("Time taken using @ operator:", end - start)

print("Results are equal:", np.allclose(matrix_multiplication, result_at))
```

```python
# 22. Question: Create a 3D array of shape (2, 1, 4) and a 2D array of shape (4, 1). Perform an element-wise operation using broadcasting and explain the result. Use np. new axis to achieve the same result without broadcasting.

A = np.array([[[1, 2, 3, 4]],
              [[5, 6, 7, 8]]])
B = np.array([[10],
              [20],
              [30],
              [40]])
print("Shape of A:", A.shape)
print("Shape of B:", B.shape)

result_broadcast = A + B
print("\nResult using broadcasting:\n", result_broadcast)
print("Result shape:", result_broadcast.shape)

B_new = B[np.newaxis, :, :]

print("Shape of B after newaxis:", B_new.shape)

result_newaxis = A + B_new
print("\nResult using np.newaxis:\n", result_newaxis)

print("\nResults are equal:",
      np.array_equal(result_broadcast, result_newaxis))
```

```python
# 24. Question: Given a 5x5 array of sequential integers, use slicing to:
"""Extract the diagonal elements
Replace the elements of the middle row with zeros
Flip the array vertically and horizontally """

A = np.arange(1,26).reshape(5,5)
print(A)

diagonal_elements = np.diag(A)
print("\n diagonal elements: ", diagonal_elements)

A[2, :] = 0
print("\n middle row: ", A)

vertical_flip = A[::-1, :]
print("\nVertically Flipped Array:\n", vertical_flip)

horizontal_flip = A[:, ::-1]
print("\nHorizontally Flipped Array:\n", horizontal_flip)
```

```python
# 28. Question: Given a 6x6 matrix, use advanced indexing and slicing to extract the upper triangular part of the matrix and set the lower triangular part to zero. Verify the result.

A = np.arange(1, 37).reshape(6, 6)
print("Original Matrix:\n", A)

upper_tri = A.copy()

lower_indices = np.tril_indices(6, k=-1)

upper_tri[lower_indices] = 0

print("\nUpper Triangular Matrix:\n", upper_tri)

is_lower_zero = np.all(upper_tri[np.tril_indices(6, k=-1)] == 0)
print("\nVerification (Lower triangular zero?):", is_lower_zero)
```
