```python
#5) Write a Python program to get the Factorial number of given numbers.

n = int(input("Enter Any number: "))

def factorial(n):
    fact = 1
    for i in range(1,n+1):
        fact *= i
    return fact

result = factorial(n)
print(f"Factorial {n} : ",result)
```

```python
#6) Write a Python program to get the Fibonacci series of given range.

n = int(input("Enter range: "))

a, b = 0, 1

fibonacci_series = []

for _ in range(n):
    fibonacci_series.append(str(a))
    a, b = b, a + b

print(f"Fibonacci_series {n} :",' '.join(fibonacci_series))
```

```python
#24) Write a Python function to insert a string in the middle of a string.

def insert_middle(main_string, insert_string):
    mid = len(main_string) // 2
    return main_string[:mid] + insert_string + main_string[mid:]

main_string = input("Enter main string: ")
insert_string = input("Enter string to insert: ")

result = insert_middle(main_string, insert_string)
print("Result:", result)

```

```python
#31) Write a Python program to count the number of strings where the string  length is 2 or more and the first and last character are same from a given list of strings.


def count_strings(strings):
    count = 0
    for s in strings:
        if len(s) >= 2 and s[0] == s[-1]:
            count += 1
    return count

list1 = ['abd', 'xyx', 'aba', '1221', 'aa']
result = count_strings(list1)

print("Number of strings:", result)
```

```python
#83) Explain Exception handling? What is an Error in Python?

"""Exception handling is a mechanism used to handle runtime errors so that the normal flow of a program is not interrupted.
In Python, exception handling is done using the try, except, else, and finally blocks.

try - code that may cause an error
except - handles the error
else - runs if no exception occurs
finally - always executes (used for cleanup)"""

"""An error in Python is a problem that occurs during the execution of a program and stops it from running normally.
Errors can occur due to invalid syntax, wrong input, or logical mistakes.
1.SyntaxError
2.ZeroDivisionError
3.TypeError
4.IndexError"""
```

```python
#84) How many except statements can a try-except block have? Name Some built-in exception classes:

"""
A try block can have multiple except statements.
This allows handling different types of exceptions separately."""

"""
Some Built-in Exception Classes in Python:

Exception
ValueError
TypeError
IndexError
KeyError
ZeroDivisionError
FileNotFoundError
IOError
NameError"""


try:
    x = int(input("Enter number: "))
    y = 100 / x
except ValueError:
    print("Invalid input")
except ZeroDivisionError:
    print("Cannot divide by zero")
except Exception:
    print("Some other error occurred")
```

```python
#85) When will the else part of try-except-else be executed?

"""
The else block is executed only when no exception occurs in the try block.
If an exception is raised - except runs, else is skipped
If no exception occurs - else runs"""

try:
    num = int(input("Enter a number: "))
    print(10 / num)
except ZeroDivisionError:
    print("Cannot divide by zero")
else:
    print("Division successful")
```

```python
#86) Can one block of except statements handle multiple exception?
"""
Yes, a single except block can handle multiple exceptions in Python by specifying them as a tuple."""

try:
    x = int(input("Enter number: "))
    y = 10 / x
except (ValueError, ZeroDivisionError):
    print("Invalid input or division by zero")
```

```python
#87) When is the finally block executed?

"""
The finally block is always executed, whether an exception occurs or not.
Runs after try and except blocks
Executes even if an error occurs or not
Used for cleanup tasks like closing files or releasing resources."""

try:
    file = open("test.txt", "r")
    print(file.read())
except FileNotFoundError:
    print("File not found")
finally:
    file.close()
    print("File closed")
```

```python
#89) How Do You Handle Exceptions with Try/Except/Finally in Python?
"""
try - Write code that may cause an error.
except - Handles the error if it occurs.
finally - Always executes, whether an exception occurs or not (used for cleanup)."""

try:
    numbers = [10, 20, 30]
    print(numbers[5])
except IndexError:
    print("Index out of range")
finally:
    print("Program finished")
```
